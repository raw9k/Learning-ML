```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred = lr.predict(x_test_scaled)
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(f"mean_squared_error = {mse}")
print(f"mean_absolute_error = {mae}")
print(f"r2_score = {r2}")
```

# Pickling
Python's pickle module is used for serializing and deserializing Python objects. Serialization (or "pickling") converts a Python object into a byte stream, which can be stored in a file or transmitted over a network. Deserialization (or "unpickling") reconstructs the original object from the byte stream.

```python
import pickle
pickle.dump(lr,open("regressor.pkl","wb"))
```

```python
Regression_model = pickle.load(open("regressor.pkl","rb"))
```

```python
Regression_model.predict(x_test_scaled)
```

```python
x = df.iloc[:,0:6]
y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=30, random_state=42)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet, RidgeCV,LassoCV,ElasticNetCV
```

```python
ridge = Ridge()
Lasso = Lasso()
ElasticNet = ElasticNet()
```

```python
ridge.fit(x_train_scaled,y_train)
Lasso.fit(x_train_scaled,y_train)
ElasticNet.fit(x_train_scaled,y_train)
```

```python
y_pred_ridge = ridge.predict(x_test_scaled)
y_pred_Lasso = Lasso.predict(x_test_scaled)
y_pred_ElasticNet = ElasticNet.predict(x_test_scaled)
```

```python
print("Results using Ridge Regression:")
mae = mean_absolute_error(y_test,y_pred_ridge)
r2 = r2_score(y_test, y_pred_ridge)
print(f"Coef of Ridge :", ridge.coef_)
print(f"mean_absolute_error = {mae}")
print(f"r2_score = {r2}")
```

```python
print("Results using Lasso Regression:")
mae = mean_absolute_error(y_test,y_pred_Lasso)
r2 = r2_score(y_test, y_pred_Lasso)
print(f"Coef of Lasso :",Lasso.coef_)
print(f"mean_absolute_error = {mae}")
print(f"r2_score = {r2}")
```

```python
print("Results using ElasticNet Regression:")
mae = mean_absolute_error(y_test,y_pred_ElasticNet)
print(f"Coef of ElasticNet :",ElasticNet.coef_)
r2 = r2_score(y_test, y_pred_ElasticNet)
print(f"mean_absolute_error = {mae}")
print(f"r2_score = {r2}")
```
