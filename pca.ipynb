```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# 1️⃣ Generate a simple 2D dataset
# ---------------------------
np.random.seed(42)
mean = [0, 0]
cov = [[3, 2], [2, 2]]  # correlated features
X = np.random.multivariate_normal(mean, cov, 100)

# ---------------------------
# 2️⃣ Center the data
# ---------------------------
X_meaned = X - np.mean(X, axis=0)

# ---------------------------
# 3️⃣ Compute Covariance Matrix
# ---------------------------
# Cov(X) = (1 / (n-1)) * XᵀX
cov_mat = np.cov(X_meaned.T)

# ---------------------------
# 4️⃣ Eigen decomposition (mathematical core)
# ---------------------------
eigen_values, eigen_vectors = np.linalg.eig(cov_mat)

# ---------------------------
# 5️⃣ Sort eigenvectors by eigenvalues (descending order)
# ---------------------------
sorted_idx = np.argsort(eigen_values)[::-1]
eigen_values = eigen_values[sorted_idx]
eigen_vectors = eigen_vectors[:, sorted_idx]

# ---------------------------
# 6️⃣ Project data onto the first principal component
# ---------------------------
projected_1D = X_meaned @ eigen_vectors[:, 0].reshape(-1, 1)

# For 2D reconstruction (back projection)
X_reconstructed = projected_1D @ eigen_vectors[:, 0].reshape(1, -1)

# ---------------------------
# 7️⃣ Visualize results
# ---------------------------

# Plot original data
plt.figure(figsize=(10, 6))
plt.scatter(X_meaned[:, 0], X_meaned[:, 1], alpha=0.6, label="Original Data")

# Plot principal component directions (orthonormal)
origin = np.mean(X_meaned, axis=0)
for i in range(2):
    vec = eigen_vectors[:, i]
    plt.quiver(origin[0], origin[1], vec[0], vec[1],
               angles='xy', scale_units='xy', scale=0.5,
               color=['r', 'g'][i], label=f'PC{i+1}')

plt.axis('equal')
plt.title("PCA from Scratch — Original Data and Principal Components")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 8️⃣ Visualize projection (data reduced to 1D and reconstructed)
# ---------------------------
plt.figure(figsize=(10, 6))
plt.scatter(X_meaned[:, 0], X_meaned[:, 1], alpha=0.4, label='Original Data')
plt.scatter(X_reconstructed[:, 0], X_reconstructed[:, 1],
            color='red', label='Projection onto 1st Principal Component')
plt.title("Projection of Data onto First Principal Component (Reconstruction)")
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# 9️⃣ Verify Orthonormality of Eigenvectors
# ---------------------------
print("Eigenvalues:\n", eigen_values)
print("\nEigenvectors (columns):\n", eigen_vectors)
print("\nCheck orthonormality (should be identity):\n", eigen_vectors.T @ eigen_vectors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
# Loading grayscale image 
img = Image.open(r"C:\Users\rauna\Downloads\image.jpg").convert('L')
X = np.array(img, dtype=float)
# Normalizing pixel values
X /= 255.0
# original
plt.figure(figsize=(6,6))
plt.imshow(X, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean
cov_matrix = np.cov(X_centered, rowvar=False)
# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)  


sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]
# number of principal components
k_values = [5, 20, 50, 100, 200]
# Reconstructing images with top-k components
plt.figure(figsize=(15,8))
plt.suptitle("PCA Image Compression (From Scratch using Orthonormal Basis)", fontsize=14)
for i, k in enumerate(k_values, 1):
    top_k_eigvecs = eigenvectors[:, :k]
    X_reduced = np.dot(X_centered, top_k_eigvecs)
    X_reconstructed = np.dot(X_reduced, top_k_eigvecs.T) + X_mean
    X_reconstructed = np.clip(X_reconstructed, 0, 1)
    plt.subplot(2, 3, i)
    plt.imshow(X_reconstructed, cmap='gray')
    plt.title(f"k = {k}")
    plt.axis('off')
plt.show()
# Checking orthonormality of eigenvectors
orth_check = np.dot(eigenvectors.T, eigenvectors)
print("Check orthonormality (should be Identity):\n", np.round(orth_check[:5,:5], 3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# ---------------------------
# 1️⃣ Load and preprocess image
# ---------------------------
# Replace with your image path
img = Image.open(r"C:\Users\rauna\Downloads\image.jpg").convert('L')
X = np.array(img, dtype=float)
X /= 255.0  # Normalize pixel values

n, m = X.shape
print(f"Original Image Shape: {n} x {m}")
print(f"Original storage (in units): {n*m}")

# Display original image
plt.figure(figsize=(6,6))
plt.imshow(X, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# ---------------------------
# 2️⃣ Center the image data
# ---------------------------
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# ---------------------------
# 3️⃣ Covariance matrix
# ---------------------------
cov_matrix = np.cov(X_centered, rowvar=False)

# ---------------------------
# 4️⃣ Eigen decomposition
# ---------------------------
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Sort descending
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]

# ---------------------------
# 5️⃣ Choose number of principal components
# ---------------------------
k_values = [5, 20, 50, 100, 200,300]

# ---------------------------
# 6️⃣ Reconstruct and compute storage
# ---------------------------
plt.figure(figsize=(15,8))
plt.suptitle("PCA Image Compression (with Storage Computation)", fontsize=14)

for i, k in enumerate(k_values, 1):
    # Top-k eigenvectors
    top_k_eigvecs = eigenvectors[:, :k]
    
    # Project to k dimensions
    X_reduced = np.dot(X_centered, top_k_eigvecs)
    
    # Reconstruct
    X_reconstructed = np.dot(X_reduced, top_k_eigvecs.T) + X_mean
    X_reconstructed = np.clip(X_reconstructed, 0, 1)
    
    # ---------------------------
    # Storage Calculation
    # ---------------------------
    original_storage = n * m
    pca_storage = k * (n + m + 1)
    compression_ratio = (pca_storage / original_storage) * 100
    
    print(f"k = {k:<4} | PCA Storage = {pca_storage:<8} | "
          f"Compression Ratio = {compression_ratio:6.2f}% of original")

    # ---------------------------
    # Visualization
    # ---------------------------
    plt.subplot(2, 3, i)
    plt.imshow(X_reconstructed, cmap='gray')
    plt.title(f"k = {k}\n({compression_ratio:.1f}% storage)")
    plt.axis('off')

plt.tight_layout()
plt.show()

# ---------------------------
# 7️⃣ Verify orthonormality of eigenvectors
# ---------------------------
orth_check = np.dot(eigenvectors.T, eigenvectors)
print("\nCheck orthonormality (should be Identity):\n", np.round(orth_check[:5, :5], 3))
```

```python
np.allclose(orth_check, np.eye(orth_check.shape[0]))
```
