```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30, random_state= 10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="linear")
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("Results using Support Vector classifier:")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf = SVC(kernel= "rbf")
```

```python
rbf.fit(x_train,y_train)
```

```python
y_rbf_pred = rbf.predict(x_test)
```

```python
print("Results using Support Vector classifier:")
print(classification_report(y_test, y_rbf_pred))
print(confusion_matrix(y_test,y_rbf_pred))
```

```python
polysv = SVC(kernel="poly")
polysv.fit(x_train,y_train)
y_poly_pred = polysv.predict(x_test)
print("Results using Support Vector classifier:")
print(classification_report(y_test, y_poly_pred))
print(confusion_matrix(y_test,y_poly_pred))
```

```python
sigmoid = SVC(kernel="sigmoid")
sigmoid.fit(x_train,y_train)
y_sig_pred = sigmoid.predict(x_test)
print("Results using Sigmoid Support Vector classifier:")
print(classification_report(y_test, y_sig_pred))
print(confusion_matrix(y_test,y_sig_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C":[0.1,1,10,100,1000],
    "gamma":[1,0.1,0.01,0.001,0.0001],
    "kernel" : ["rbf","poly"],
    "degree" : [2,3,4,5]
}
```

```python
grid = GridSearchCV(SVC(), param_grid= param_grid, cv=5, verbose=3, refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
y_grid_pred = grid.predict(x_test)
print("Results using Sigmoid Support Vector classifier:")
print(classification_report(y_test, y_grid_pred))
print(confusion_matrix(y_test,y_grid_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

# Generating a toy dataset with 2 features
X, y = make_blobs(n_samples=100, centers=2, random_state=6)

# Training a linear SVM
clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, y)
# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='autumn')

# Getting the hyperplane separated
w = clf.coef_[0]
b = clf.intercept_[0]

# Plotting decision boundary: wÂ·x + b = 0
x_points = np.linspace(min(X[:, 0]), max(X[:, 0]))
y_points = -(w[0] * x_points + b) / w[1]
plt.plot(x_points, y_points, 'k-', label='Decision boundary')

# Plotting margin lines: wÂ·x + b = Â±1
margin = 1 / np.linalg.norm(w)
y_margin_pos = -(w[0] * x_points + b - 1) / w[1]
y_margin_neg = -(w[0] * x_points + b + 1) / w[1]

plt.plot(x_points, y_margin_pos, 'k--', label='+1 margin')
plt.plot(x_points, y_margin_neg, 'k--', label='-1 margin')

# Plotting support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='blue', label='Support Vectors')

plt.legend()
plt.title("SVM Decision Boundary with Margins")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC

# Step 1: Generate the dataset
x, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_classes=2,
    n_clusters_per_class=1,
    n_redundant=0,
    random_state=8
)

# Step 2: Train a linear SVM classifier
clf = grid.best_estimator_
clf.fit(x, y)

# Step 3: Visualize decision boundary and margins
# Create a mesh grid
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Evaluate decision function over the grid
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, levels=20, cmap='coolwarm', alpha=0.2)  # background shading
plt.contour(xx, yy, Z, levels=[-1, 0, 1], colors=['blue', 'black', 'red'], linestyles=['--', '-', '--'])

# Scatter the data points
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='coolwarm', s=30, edgecolors='k')

# Highlight support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='k', label='Support Vectors')

plt.title("SVC with Linear Kernel: Decision Boundary and Margins")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
