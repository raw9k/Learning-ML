```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=3,n_redundant=1,
                          n_classes=2,random_state=999)
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=8,weights="distance",
                algorithm="auto")
knn_classifier.fit(x_train,y_train)
```

```python
y_pred = knn_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
acc = accuracy_score(y_test,y_pred)
confusion_matrix = confusion_matrix(y_test,y_pred)
classification_report = classification_report(y_test,y_pred)
print(f"accuracy_score = {acc}")
print(f"confusion_matrix = {confusion_matrix}")
print(f"classification_report = {classification_report}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_neighbors":[5,8,10,13,16,19],
    "weights":["distance", "uniform"],
    "algorithm": ['auto', 'ball_tree', 'kd_tree'],
    "p":[0,1]
}
```

```python
grid = GridSearchCV(KNeighborsClassifier(),param_grid= param_grid, refit=True,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
confusion_matrix = confusion_matrix(y_test,y_pred)
classification_report = classification_report(y_test,y_pred)
print(f"accuracy_score = {acc}")
print(f"confusion_matrix = {confusion_matrix}")
print(f"classification_report = {classification_report}")
```
